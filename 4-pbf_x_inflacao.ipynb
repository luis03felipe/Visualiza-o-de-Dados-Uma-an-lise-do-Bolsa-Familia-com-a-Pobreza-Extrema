```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
```

# --==Bolsa Família==--

```python
bolsa_familia = pd.read_csv('distribuicao_bolsafamilia.csv')
bolsa_familia
```

```python
bolsa_familia = bolsa_familia.drop(['_id', 'co_uf', 'variavel', 'social_categoria', ], axis=1) 
bolsa_familia
```

```python
data = {'Estado': bolsa_familia['no_uf'],
        'Ano': bolsa_familia['ano'],
        'Categoria': bolsa_familia ['social_subcategoria'],
        'Valor médio em R$': bolsa_familia ['valor']
        }
```

```python
dataFrame = pd.DataFrame(data)
dataFrame
```

```python
df_filtered = dataFrame[dataFrame.index % 3 == 0] #linhas que são multiplas de 3
df_filtered.reset_index(drop=True, inplace=True)
df_filtered
```

```python
nova_ordem = ['Ano', 'Estado', 'Valor médio em R$', 'Categoria' ]
df = df_filtered[nova_ordem]
df
```

```python
df = df.sort_values(by=['Estado', 'Ano'], ascending=True)#para colocar em ordem alfabetica
df = df.drop(['Categoria', 'Estado'], axis=1) 
df
```

```python
bolsa_familia_anual = df.groupby('Ano').mean().reset_index()
bolsa_familia_anual
```

```python
pd.set_option('float_format', '{:.2f}'.format) #formatação para numeros cientificos
bolsa_familia_anual
```

# --==Inflação==--

```python
inflacao = pd.read_csv('inflacao.csv')
inflacao
```

```python
pd.set_option('display.max_columns', None)
inflacao
```

```python
inflacao = inflacao.drop(['ipca_acumulado_doze_meses', 
                          'ipca_variacao',
                          'referencia',
                          'mes',
                          'ano_mes',
                          'ipca15_variacao',
                          'ipca15_acumulado_ano',
                          'ipca15_acumulado_doze_meses',
                          'inpc_variacao',
                          'inpc_acumulado_ano',
                          'inpc_acumulado_doze_meses',
                          'ipa_variacao',
                          'ipa_acumulado_ano',
                          'ipc_fipe_variacao',
                          'ipc_fipe_acumulado_ano',
                          'incc_variacao',
                          'incc_acumulado_ano', 
                          'incc_m_variacao',
                          'incc_m_acumulado_ano', 
                          'selic_meta', 
                          'selic_ano', 
                          'juros_reais', 
                          'salario_minimo', 
                          'consolidado_ano' 
                          ], axis=1) 
inflacao
```

```python
inflacao_att = inflacao.groupby('Ano').mean().reset_index()
inflacao_att
```

```python
inflacao_att.info()
```

```python
inflacao_att.index = inflacao_att.index.astype(int)
inflacao_att.info()
```

```python
inflacao_anual = {'Ano': inflacao_att['Ano'],
                  'IPCA': inflacao_att['ipca_acumulado_ano']
}
inflacao_anual
```

# --==Bolsa Família X Inflação==--

```python
bolsa_familia_anual['Ano'] = pd.to_datetime(bolsa_familia_anual['Ano'], format='%Y')
inflacao_anual['Ano'] = pd.to_datetime(inflacao_anual['Ano'], format='%Y')
```

```python
if isinstance(bolsa_familia_anual, dict):
    bolsa_familia_anual = pd.DataFrame(bolsa_familia_anual)

if isinstance(inflacao_anual, dict):
    inflacao_anual = pd.DataFrame(inflacao_anual)
```

```python
df_merged = pd.merge(bolsa_familia_anual, inflacao_anual, on='Ano')

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(df_merged['Ano'], df_merged['Valor médio em R$'], marker='o', linestyle='-', color='b', label='Média do Bolsa Família (R$)')
ax1.set_xlabel('Ano')
ax1.set_ylabel('Média do Bolsa Família (R$)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
ax2.plot(df_merged['Ano'], df_merged['IPCA'], marker='o', linestyle='-', color='r', label='IPCA ao Ano(%)')
ax2.set_ylabel('IPCA (%)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

plt.title('Média Anual do Bolsa Família e IPCA acumulado ao Ano')
ax1.grid(True)
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

plt.show()
```
