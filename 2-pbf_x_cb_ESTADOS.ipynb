```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
```

# --==Distribuição do Programa Bolsa Família==--

```python
bolsa_familia = pd.read_csv('distribuicao_bolsafamilia.csv')
bolsa_familia
```

```python
bolsa_familia = bolsa_familia.drop(['_id', 'co_uf', 'variavel', 'social_categoria', ], axis=1) 
bolsa_familia
```

```python
data = {'Estado': bolsa_familia['no_uf'],
        'Ano': bolsa_familia['ano'],
        'Categoria': bolsa_familia ['social_subcategoria'],
        'Valor médio em R$': bolsa_familia ['valor']
        }
```

```python
dataFrame = pd.DataFrame(data)
dataFrame
```

```python
df_filtered = dataFrame[dataFrame.index % 3 == 0] #linhas que são multiplas de 3
df_filtered.reset_index(drop=True, inplace=True)
df_filtered
```

```python
pd.set_option('float_format', '{:.2f}'.format) #formatação para numeros cientificos
df_filtered
```

```python
nova_ordem = ['Ano', 'Estado', 'Valor médio em R$', 'Categoria' ]
df = df_filtered[nova_ordem]
df
```

```python
df = df.sort_values(by=['Estado', 'Ano'], ascending=True)#para colocar em ordem alfabetica
df
```

```python
df = df.drop(['Categoria', 'Ano'], axis=1) 
df
```

```python
plt.figure(figsize=(16, 10))
sns.histplot(data= df, x='Estado', y= 'Valor médio em R$', bins=10, kde=True)
plt.show()
```

```python
df_anual = df.groupby('Estado').mean().reset_index()
df_anual
```

```python
df_anual = df_anual.sort_values(by='Valor médio em R$', ascending=False)
df_anual
plt.figure(figsize=(12, 6))
sns.barplot(x='Estado', y='Valor médio em R$', data=df_anual, errorbar=None, estimator=np.mean)
plt.xlabel('Estado')
plt.ylabel('Valor médio em R$')
plt.title('Média do benefício recebido pelas famílias do Bolsa Família por estado (2004-2023)')
plt.xticks(rotation=0)  # Rotacionar os rótulos do eixo x se necessário
plt.show()
```

```python
cesta_basica = pd.read_excel('cestas_basicas.xls')
cesta_basica
```

```python
cb_anual = cesta_basica.groupby('Ano').mean().reset_index()
cb_anual
```

```python
df_organizado = cb_anual.reindex(sorted(cb_anual.columns), axis=1)
df_organizado
```

```python
coluna_ano = df_organizado.pop('Ano')
df_organizado.insert(0, 'Ano', coluna_ano)
df_organizado
```

```python
linha_2022 = df_organizado[df_organizado['Ano'] == 2022]
soma_2022_todas_colunas = linha_2022.drop(columns=['Ano']).sum(axis=1).values[0]
soma_2022_todas_colunas
```

```python
df_sem_nan = df_organizado.fillna(0)
df_sem_nan
```

```python
def somar_todas_colunas(linha):
    return linha.drop(labels='Ano').sum()
df_sem_nan['Soma'] = df_sem_nan.apply(somar_todas_colunas, axis=1)
df_sem_nan
```

```python
def somar_valores(linha):
    soma = 0
    estado = 0
    for valor in linha:
        if valor != 0:
            soma += valor
            estado += 1
    return soma / estado if estado != 0 else 0
# Aplicar a função a todas as linhas e criar uma nova coluna 'Resultado'
df_sem_nan['Resultado'] = df_sem_nan.apply(somar_valores, axis=1)
df_sem_nan
```

```python
df = df_sem_nan.set_index('Ano')
df.index = df.index.astype(int)
df
```

```python
df.info()
```

```python
df.index = df.index.astype(int) # Certificar que o índice é inteiro

plt.figure(figsize=(15, 10))

# Plotar os dados
plt.plot(df.index, df['Resultado'], marker='o', linestyle='-', color='g')
plt.xlabel('Ano')
plt.ylabel('Valor da cesta (em R$)')
plt.title('Cestas básicas: Média de todos os estados x Ano')

plt.xticks(ticks=df.index) # Ajustar os ticks do eixo x para mostrar apenas os anos

plt.yticks(df['Resultado'])# Ajustar os ticks do eixo y para os valores presentes em 'Resultado'

plt.show()
```
